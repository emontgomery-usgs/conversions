```python
import netCDF4 as nc
from netCDF4 import Dataset
import os
```

```python
# This program was used to traverse all the directories containing EPIC data and 
# replace any MOORING attributes that have an extra digit with the right value, as a string
# first work in a test area
#root_path="/usgs/Projects/stellwagen/Data/tst"
# then unleash on the real data
root_path="/usgs/Projects/stellwagen/Data"
dirs=os.listdir(root_path)
```

```python
# everything below here was for testing/verification
```

```python
for dirs in os.listdir(root_path):
    dname=os.path.join(root_path, dirs)
    for filename in os.listdir(dname):
        if filename.endswith(".nc") or filename.endswith(".cdf"):
            fname=os.path.join(dname, filename)
            fnc=nc.Dataset(fname,mode='r+',format='NETCDF3_CLASSIC') # don't use 'w' will delete existing file!
            try:
                m_att=fnc.getncattr('MOORING')
            except:
                print(" ".join(['No MOORING att in', filename, 'getting from filename']))
                m_att=filename[:3]
            # if m_att is a float, str(m_att) returns '8551.0', 
            # use str(int(m_att)) to get '8551'
            if not isinstance(m_att, str):
                m_att=str(int(m_att))           
            #print(" ".join([filename, ": MOORING att is: ", m_att]))
            fparts=fname.rpartition('/') # use right partition to get at the least element easily
            midname=fparts[-1]
            #if len(m_att) == 4 and filename.startswith("10"):
                #print('length = 4; do nothing')
            if len(m_att) == 4 and not filename.startswith("10"):
                if int(midname[:3]) <= 999:
                    print(" ".join([filename, ": changing MOORING att to: ", midname[:3]]))
                    fnc.setncattr('MOORING',midname[:3])
                #else:
                    #print ('length = 4')
            elif len(m_att) == 5 and filename.startswith("10"):
                fnc.setncattr('MOORING',midname[:4])
                print(" ".join([filename, ": changing MOORING att to: ", midname[:4]]))
            #elif len(m_att) == 3:
                #print('length = 3; do nothing')
            #else:
                #print(" ".join([filename, ": length not 3, 4 or 5"]))
            fnc.close()
    print(" ".join([dirs, ": end of correcting MOORING"]))
```

```python
 # check there's nothing wrong in ECOHAB_II
dname=os.path.join(root_path, 'ECOHAB_II')
for filename in os.listdir(dname):
    if filename.endswith(".nc") or filename.endswith(".cdf"):
        fname=os.path.join(dname, filename)
        fnc=nc.Dataset(fname,mode='r+',format='NETCDF3_CLASSIC') # don't use 'w' will delete existing file!
        m_att=fnc.getncattr('MOORING')
        print(" ".join([filename, ": MOORING att is: ", m_att]))
        fnc.close()
print('end of ECOHAB_II check')    
  
```
